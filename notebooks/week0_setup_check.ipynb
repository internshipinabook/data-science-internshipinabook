```python
# Check that every library needed for this course is installed
libraries = [
    ('pandas',      'Data manipulation — DataFrames, loading CSVs'),
    ('numpy',       'Numerical computing — arrays, maths'),
    ('matplotlib',  'Plotting — charts and visualisations'),
    ('seaborn',     'Statistical visualisation — built on matplotlib'),
    ('sklearn',     'Machine learning — models, metrics, preprocessing'),
    ('streamlit',   'Dashboard deployment — Week 10'),
]

all_ok = True
for lib, description in libraries:
    try:
        module = __import__(lib)
        version = getattr(module, '__version__', 'installed')
        print(f"✅ {lib:<15} {version:<10}  {description}")
    except ImportError:
        print(f"❌ {lib:<15} NOT FOUND   {description}")
        print(f"   Fix: pip install {lib}")
        all_ok = False

print()
if all_ok:
    print("✅ All libraries installed correctly.")
else:
    print("❌ Install missing libraries above, then re-run this cell.")
```

## Step 5 — Check Dataset Files

```python
import pandas as pd, os

# Smart loader — tries local package first, then GitHub, then generates sample data
LOCAL_PATH = '../data/supermarket_sales.csv'
REMOTE_URL = 'https://raw.githubusercontent.com/internshipinabook/data-science-internshipinabook/main/data/supermarket_sales.csv'

df = None
if os.path.exists(LOCAL_PATH):
    df = pd.read_csv(LOCAL_PATH)
    print(f'Loaded from local package ✅')
else:
    try:
        df = pd.read_csv(REMOTE_URL)
        print(f'Loaded from GitHub repository ✅')
    except Exception as e:
        print(f'⚠️  Dataset not yet available online ({e})')
        print('Generating sample data for now — upload supermarket_sales.csv to the data/ folder')
        import numpy as np; np.random.seed(42)
        df = pd.DataFrame({
            'Invoice ID': [f'INV-{i:04d}' for i in range(100)],
            'Branch': np.random.choice(['A','B','C'], 100),
            'City': np.random.choice(['Yangon','Mandalay','Naypyitaw'], 100),
            'Customer type': np.random.choice(['Member','Normal'], 100),
            'Gender': np.random.choice(['Male','Female'], 100),
            'Product line': np.random.choice(['Health and beauty','Electronic accessories','Food and beverages'], 100),
            'Unit price': np.round(np.random.uniform(10,100,100),2),
            'Quantity': np.random.randint(1,11,100),
            'Total': np.round(np.random.uniform(50,500,100),2),
            'Rating': np.round(np.random.uniform(4,10,100),1),
        })
        print('Sample dataset ready (100 rows) ✅')

print(f'Shape: {df.shape[0]:,} rows × {df.shape[1]} columns')
df.head()
```

## Step 6 — Quick Test: Load and Plot

```python
# A quick end-to-end check that pandas and matplotlib work together
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configure matplotlib for inline display in Jupyter
%matplotlib inline
sns.set_theme(style='whitegrid')

# Create a tiny test DataFrame — no real data needed for this test
test_df = pd.DataFrame({
    'week':  list(range(1, 13)),
    'skills': [i**1.4 for i in range(1, 13)]
})

# Plot it
fig, ax = plt.subplots(figsize=(7, 3))
ax.fill_between(test_df['week'], test_df['skills'],
                color='#2E75B6', alpha=0.3)
ax.plot(test_df['week'], test_df['skills'],
        color='#1A3A5C', linewidth=2, marker='o', markersize=4)
ax.set_title('Your skills after 12 weeks — projected', fontweight='bold', loc='left')
ax.set_xlabel('Week'); ax.set_ylabel('Capability (arbitrary units)')
ax.set_xticks(range(1, 13))
ax.spines['top'].set_visible(False); ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.show()

print("✅ pandas, numpy, matplotlib, seaborn — all working.")
```
