```python
import matplotlib.pyplot as plt, matplotlib.patches as mpatches

def make_exec_slide(title_text, subtitle='Kova Analytics | Churn Pilot Review'):
    fig = plt.figure(figsize=(13.33, 7.5)); fig.patch.set_facecolor('#FFFFFF')
    header = fig.add_axes([0, 0.86, 1, 0.14]); header.set_facecolor('#1A3A5C'); header.axis('off')
    header.text(0.04, 0.6, subtitle, color='white', fontsize=9.5, va='center', transform=header.transAxes)
    content = fig.add_axes([0.05, 0.05, 0.9, 0.77]); content.axis('off')
    content.text(0.0, 0.95, title_text, fontsize=18, fontweight='bold', color='#1A3A5C',
                 transform=content.transAxes, va='top')
    return fig, content
print('✅ make_exec_slide() defined')
```
