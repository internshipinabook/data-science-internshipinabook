## ⚠️ Common Mistakes

| Mistake | What goes wrong | Fix |
|---------|-----------------|-----|
| Jumping to model fitting without a train/test split | If you train and evaluate on the same data, every model looks perfect. Split before you touch the data. | Always `train_test_split` before any `.fit()` call |
| Treating accuracy as the only metric | On imbalanced data, a model that predicts 'no churn' for everyone scores 85% accuracy. That model is useless. | Always check precision, recall, and F1 alongside accuracy |
